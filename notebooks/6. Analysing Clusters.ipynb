# Vamos a analizar los diferentes clusters para ver qué los identifica

```python
import pandas as pd
```

```python
df = pd.read_csv('../data2/processed/FINAL_CLUSTERED.csv', index_col = 0)
```

```python
df = df.drop(columns=('vat'))
```

```python
df.describe()
```

```python
df['clusters'].value_counts()
```

```python
for i in range(-1,6):
    print (f'cluster: {i}', df[df['clusters']==i]['total_value'].describe())
```

```python
for i in range(-1,6):
    print (f'cluster: {i}', df[df['clusters']==i]['provincia'].value_counts())
```
