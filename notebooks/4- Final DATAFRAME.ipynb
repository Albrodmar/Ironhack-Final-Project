```python
import pandas as pd
```

```python
df = pd.read_csv('../data2/processed/merged_clean.csv', index_col = 0)
```

```python
df['value'] = df['value'].astype(float)
df['quantity'] = df['quantity'].astype(int)
```

```python
total_values = df.groupby('vat')['value'].sum()
total_quantity = df.groupby('vat')['quantity'].sum()
```

```python
df.head()
```

```python
df1 = pd.merge(df, total_values, on='vat')
```

```python
df1.head()
```

```python
aggresults = pd.merge(df1, total_quantity, on='vat')
```

```python
aggresults.rename(columns={'value_x':'value','quantity_x':'quantity','value_y':'total_value', 'quantity_y':'total_quantity'}, inplace = True)
```

```python
aggresults['category'].value_counts()
```

```python
aggresults['Wristband_value'] = aggresults[aggresults['category'] == 'Wristband'].value
aggresults['Wristband_quantity'] = aggresults[aggresults['category'] == 'Wristband'].quantity

aggresults['RFID_value'] = aggresults[aggresults['category'] == 'RFID'].value
aggresults['RFID_quantity'] = aggresults[aggresults['category'] == 'RFID'].quantity

aggresults['Tickets_value'] = aggresults[aggresults['category'] == 'Tickets'].value
aggresults['Tickets_quantity'] = aggresults[aggresults['category'] == 'Tickets'].quantity

aggresults['CardPrinter_value'] = aggresults[aggresults['category'] == 'Card Printer'].value
aggresults['CardPrinter_quantity'] = aggresults[aggresults['category'] == 'Card Printer'].quantity

aggresults['TouchScreen_value'] = aggresults[aggresults['category'] == 'Touch Screen & Computer'].value
aggresults['TouchScreen_quantity'] = aggresults[aggresults['category'] == 'Touch Screen & Computer'].quantity

aggresults['Card_value'] = aggresults[aggresults['category'] == 'Card'].value
aggresults['Card_quantity'] = aggresults[aggresults['category'] == 'Card'].quantity

aggresults['CardRibbon_value'] = aggresults[aggresults['category'] == 'Card Ribbon'].value
aggresults['CardRibbon_quantity'] = aggresults[aggresults['category'] == 'Card Ribbon'].quantity

aggresults['Mobile_value'] = aggresults[aggresults['category'] == 'Mobile Computing'].value
aggresults['Mobile_quantity'] = aggresults[aggresults['category'] == 'Mobile Computing'].quantity

aggresults['Scanner_value'] = aggresults[aggresults['category'] == 'Scanner'].value
aggresults['Scanner_quantity'] = aggresults[aggresults['category'] == 'Scanner'].quantity

aggresults['LabelPrinter_value'] = aggresults[aggresults['category'] == 'Label Printer'].value
aggresults['LabelPrinter_quantity'] = aggresults[aggresults['category'] == 'Label Printer'].quantity

aggresults['Label_value'] = aggresults[aggresults['category'] == 'Label'].value
aggresults['Label_quantity'] = aggresults[aggresults['category'] == 'Label'].quantity

aggresults['Ribbon_value'] = aggresults[aggresults['category'] == 'Ribbon'].value
aggresults['Ribbon_quantity'] = aggresults[aggresults['category'] == 'Ribbon'].quantity

aggresults['TicketPrinter_value'] = aggresults[aggresults['category'] == 'Ticket Printer'].value
aggresults['TicketPrinter_quantity'] = aggresults[aggresults['category'] == 'Ticket Printer'].quantity

aggresults['Printhead_value'] = aggresults[aggresults['category'] == 'Printhead'].value
aggresults['Printhead_quantity'] = aggresults[aggresults['category'] == 'Printhead'].quantity

aggresults['Accessories_value'] = aggresults[aggresults['category'] == 'Accessories'].value
aggresults['Accessories_quantity'] = aggresults[aggresults['category'] == 'Accessories'].quantity
```

```python
aggresults.drop(columns=['category','value','quantity'], inplace = True)
```
