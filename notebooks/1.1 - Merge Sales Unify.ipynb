```python
import pandas as pd
```

```python
df = pd.read_csv('../data2/processed/merged_clean.csv', index_col = 0)
```

```python
df
```

```python
df['category'].value_counts()
```

```python
cat_list = list(df['category'].unique())
cat_list
```

```python
for x in cat_list:
    df[x+'_value'] = df[df['category'] == x]['value']
    df[x+'_qty'] =  df[df['category'] == x]['quantity']
```

```python
df
```

```python
df['vat'].value_counts()
```

```python
df = df.drop_duplicates('vat')
```

```python
df['vat'].value_counts()
```

```python
df = df.fillna(0)
```
