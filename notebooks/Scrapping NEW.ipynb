```python
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.webdriver.chrome.options import Options
```

```python
URL = 'https://www.einforma.com/servlet/app/prod/ETIQUETA_EMPRESA/nif/'
```

```python
#cifs = pd.read_excel('new_cifs_con_Nombre.xlsx', index_col=0)
cifs = pd.read_csv('..\Datos\processed\sales_clean.csv', index_col=0)
URL = 'https://www.einforma.com/servlet/app/prod/ETIQUETA_EMPRESA/nif/'
nan_value = float("NaN")
cifs['Localidad'] = nan_value
cifs['Actividad Informa'] = nan_value
cifs['Objeto Social'] = nan_value
cifs['Forma Juridica'] = nan_value
```

```python
cifs = cifs.reset_index()
```

```python
cifs.head()
```

```python
#Chrome User Agent
options = Options()
options.add_argument("window-size=800,800")
from fake_useragent import UserAgent
ua = UserAgent()
a = ua.random
user_agent = ua.random
print(user_agent)
options.add_argument(f'user-agent={user_agent}')
driver = webdriver.Chrome(options=options)
```

```python
total_cif = 0
total_localidad = 0
total_actividad = 0
total_objeto = 0
total_forma = 0
bucle = 0
no_disponible = 0
reinicio = 0



for i in range(len(cifs)):
    bucle += 1
    print(f'bucle: {bucle}')
    driver.get(URL+ cifs.iloc[i]['vat'])
    time.sleep(0)
    try:
        driver.find_element_by_xpath('//*[@id="content"]/div/div[1]/p').text == 'Servicio No Disponible'
        reinicio += 1
        print(f'Reinicios de navegador: {reinicio}')
        driver.close()
        time.sleep(2)
        user_agent = ua.random
        options.add_argument(f'user-agent={user_agent}')
        driver = webdriver.Chrome(options=options)
        
    except:

        try:
            tabla = driver.find_element_by_class_name('ficha-empresa').text
            tabla = tabla.split('\n')
            for s in range(len(tabla)):
                tabla[s] = tabla[s].split(':')
             
            for f in range(len(tabla)):
                 if tabla[f][0] == 'CNAE 2009':
                    #cifs.iloc[i]['CNAE'] = tabla[f][1]
                    cifs.loc[i,'vat'] = (tabla[f][1])
                    total_cif +=1
                    #print(f'total cif = {total_cif}')
                               
            for f in range(len(tabla)):
                 if tabla[f][0] == 'Localidad':
                    #cifs.iloc[i]['Localidad'] = (tabla[f][1])
                    cifs.loc[i,'Localidad'] = (tabla[f][1])
                    total_localidad +=1
                    #print(f'total loc = {total_localidad}')
                    
            for f in range(len(tabla)):
                 if tabla[f][0] == 'Actividad Informa':
                    #cifs.iloc[i]['Actividad Informa'] = (tabla[f][1])
                    cifs.loc[i,'Actividad Informa'] = (tabla[f][1])
                    total_actividad +=1
                    #print(f'total act = {total_actividad}')
                    
            for f in range(len(tabla)):
                 if tabla[f][0] == 'Objeto Social':
                    #cifs.iloc[i]['Actividad Informa'] = (tabla[f][1])
                    cifs.loc[i,'Objeto Social'] = (tabla[f][1])
                    total_objeto +=1
                    #print(f'total obj = {total_objeto}')                   
                    
            for f in range(len(tabla)):
                 if tabla[f][0] == 'Forma JurÃ­dica':
                    #cifs.iloc[i]['Actividad Informa'] = (tabla[f][1])
                    cifs.loc[i,'Forma Juridica'] = (tabla[f][1])
                    total_forma +=1
                    #print(f'total form = {total_forma}')  
                                   
        except:
            no_disponible += 1
            print(f'Cuenta no encontrado: {no_disponible}')
            pass
```

```python
cifs.to_csv('..\Datos\processed\cifs_clean.csv')
```
